```python
%%file calc_functions.py

def factorial(n):
    n=int(n)
    ans=n
    while n>1:
        ans=ans*(n-1)
        n=n-1
    return ans

def square(n):
    n=int(n)
    return n**2

def sqrt(n):
    n=int(n)
    return n**0.5

def from_binary(n):
    return int(n,2)
```

```python
#계산기 완성

from tkinter import *
import calc_functions
from decimal import *

#숫자 버튼
def click(key):
    if key == '=':
        try:
            result=str(eval(display.get()))[0:10]
            display.insert(END,'=' + result)
        except:
            display.insert(END,'-->Error!')
        
    elif key == 'C':
        display.delete(0,END)
        
    elif key == functions_list1[0]:
        display.delete(0,END)
        display.insert(END, "3.141592654")
        
    elif key==functions_list1[1]:
        display.delete(0,END)
        display.insert(END,'300000000')
        
    elif key==functions_list1[2]:
        display.delete(0,END)
        display.insert(END,'330')
        
    elif key==functions_list1[3]:
        display.delete(0,END)
        display.insert(END,'149597887.5')
        
    elif key==functions_list[0]:
        n=display.get()
        display.delete(0,END)
        display.insert(END,calc_functions.factorial(n) )
        
    elif key==functions_list[1]:
        n=display.get()
        display.delete(0,END)
        display.insert(END,calc_functions.square(n) )
            
        
    elif key==functions_list[2]:
        n=display.get()
        display.delete(0,END)
        display.insert(END,calc_functions.sqrt(n) )
        
        
    elif key==functions_list[3]:
        n=display.get()
        display.delete(0,END)
        display.insert(END,calc_functions.from_binary(n) )
                
        
    else:
        display.insert(END, key)
        
#메인
window=Tk()
window.title("cal")

top_row=Frame(window)
top_row.grid(row=0,column=0,columnspan=2, sticky=N)

#내용수정이 가능한 엔트리 위젯을 사용해 결과 디스플레이 사용
display=Entry(top_row, width=45, bg="light green")
display.grid()

num_pad=Frame(window)
num_pad.grid(row=1,column=0, sticky=W)

num_pad_list=[
    '7','8','9',
    '4','5','6',
    '1','2','3',
    '0','.','='
]

r=0
c=0

for btn_text in num_pad_list:
    def cmd(x=btn_text):
        click(x)
    Button(num_pad, text=btn_text, width=5, command=cmd).grid(row=r,column=c)
    c=c+1
    if c>2:
        c=0
        r=r+1
        
        

plus_pad=Frame(window)
plus_pad.grid(row=1,column=1, sticky=E)

plus_pad_list=[
    '*','/',
    '+','-',
    '(',')',
    'C'
]

r=0
c=0

for btn_text1 in plus_pad_list:
    def cmd1(x=btn_text1):
        click(x)
    Button(plus_pad, text=btn_text1, width=5, command=cmd1).grid(row=r,column=c)
    c=c+1
    if c>1:
        c=0
        r=r+1
        
functions = Frame(window)
functions.grid(row=2, column=1, sticky=E)

functions_list=[
    'factorial(!)',
    'square',
    'sqrt(square root)',
    'binary -> 10'
]

r=0
c=0
for btn_text in functions_list:
    def cmd(x=btn_text):
        click(x)
    Button(functions, text=btn_text, width=13, command=cmd).grid(row=r, column=c)
    r=r+1
        

functions1 = Frame(window)
functions1.grid(row=2, column=0, sticky=W)

functions_list1=[
    'pi',
    '빛의 이동 속도 (m/s)',
    '소리의 이동 속도 (m/s)',
    '태양과의 평균거리 (km)'
]

r=0
c=0
for btn_text in functions_list1:
    def cmd(x=btn_text):
        click(x)
    Button(functions1, text=btn_text, width=20, command=cmd).grid(row=r, column=c)
    r=r+1
        
        
#실행
window.mainloop()
```
