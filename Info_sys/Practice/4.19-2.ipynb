```python
from tkinter import*
import random

def click():
    entered_text=entry.get()
    output.delete(0.0, END)
    try:
        answer=my_glossary[entered_text]
    except:
        answer="단어의 정의를 찾을 수 없습니다."
    output.insert(END, answer)
    
def click1():
    questions=list(my_glossary.keys())
    question=random.choice(questions)
    entry.delete(0, END)
    output.delete(0.0, END)
    entry.insert(END, question)
    
window=Tk()
window.title("my korean vocab app")

Button(window, text="다음 질문", width=10, command=click1). grid(row=0, column=0, sticky=W)

entry=Entry(window, width=20,bg="light green")
entry.grid(row=1, column=0,columnspan=2, sticky=W)

Button(window, text="정답얻기", width=10, command=click). grid(row=3, column=0, sticky=W)

Label(window, text="\n 정답:").grid(row=5, column=0, sticky=W)

output=Text(window, width=75, height=6,wrap=WORD, background="light green")
output.grid(row=6, column=0, columnspan=2,sticky=W)

my_glossary={
    'school':'학교',
    'cat':'고양이',
    'bridge':'다리',
    'ghost':'유령',
    'house':'집',
    'jisu':'사랑'
}

window.mainloop()
```
